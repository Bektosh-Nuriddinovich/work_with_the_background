```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

# Image+Image

```python
# Paths
img1_path="images/pizza.jpg"
img2_path="images/sky.jpg"
```

```python
# read images
img1=cv2.imread(img1_path)
img2=cv2.imread(img2_path)

img1=cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)

img2=cv2.resize(img2,(img1.shape[1],img1.shape[0]))

plt.figure(figsize=(12,3))

plt.subplot(1,2,1)
plt.title("Image1")
plt.imshow(img1)

plt.subplot(1,2,2)
plt.title("Image2")
plt.imshow(img2)
plt.show()
```

```python
print(img1.shape)
print(img2.shape)
```

```python
# make to mask

low=np.array([0,0,230])
high=np.array([230,230,255])

mask=cv2.inRange(img1,low,high)

plt.title("Mask")
plt.imshow(mask,cmap="gray")
plt.show()
```

```python
# Cropping img1 by mask

img1_copy=np.copy(img1)
img1_copy[mask!=0]=[0,0,0]
plt.imshow(img1_copy)
```

```python
# Cropping img2 by mask

img2_copy=np.copy(img2)
img2_copy[mask == 0]=[0,0,0]
plt.imshow(img2_copy)
plt.show()
```

```python
result=img1_copy+img2_copy

plt.figure(figsize=(12,3))
plt.title("result")
plt.imshow(result)
plt.show()
```
